```python
numeric_dataSet = data_set.select_dtypes(include=["number"])
```

```python
correlation = numeric_dataSet.corr()
```

```python
colorPattern = sns.diverging_palette(20, 220, as_cmap=True)
colorPattern
```

```python
plt.figure(figsize=(8, 8))
```

```python
sns.heatmap(correlation,  annot=False, cmap=colorPattern,  cbar=True)
plt.title("Heatmap of Correlation ")
plt.show()
```

Intepretation of CO2 Diagram

Frequency analysis:

Effect of Bin Size:

When we use fewer bins (eg: bins=10), the graph looks smoother and gives a general overview of the shape.

When we use more bins (eg: bins=100), we can spot smaller patterns or groupings in the data.

Overall Observation:
Most vehicles in the dataset have moderate CO emissions. There are very few outliers, which makes the distribution centered and easy to analyze.
